```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

**Loading the dataset in a pandas dataframe**

```python
credit_card_data = pd.read_csv('/content/creditcard.csv')
```

```python
#printing the first five rows of the dataset
credit_card_data.head()
```

```python
#printing the last five rows of the dataset
credit_card_data.tail()
```

```python
#checking the number of missing values in each column
credit_card_data.isnull().sum()
```

```python
#distribution of legit transactions & fraudulent trasactions
credit_card_data['Class'].value_counts()
```

```python
#we have to perform processing for handling the data imbalance
```

**Separating the data for analysis**

```python
#creating two variables to store
#using pandas series data structure
legit = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]

print(legit.shape)
print(fraud.shape)
```

```python
#statistical measures of the data
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
#Comparing the values for both the transactions
#using groupby function
credit_card_data.groupby('Class').mean()
```

### **Performing under-sampling**
##### since the data is highly imbalanced, so we perform under sampling that is building a sample dataset containing similar distribution of legit and fraudulent transactions

```python
legit_sample = legit.sample(n=492)
#it will take random data points..random sampling

#concatenating two dataframes

new_dataset = pd.concat([legit_sample, fraud], axis=0)
#axis=0..values will be added one by one..row-wise
#axis=1..values will be added column-wise
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

```python
#in case of bad sample..the mean values will be very different
```

**Splitting** **the** **data** **into** **features** **and** **targets**

```python
X = new_dataset.drop(columns='Class', axis=1)
Y = new_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

### **Splitting the data into training and testing data**

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
#stratify makes sure that the distribution of train and test data is even
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

### **Training the Model**
#### Using Logistic Regression

```python
model = LogisticRegression()
```

```python
#training the Logistic Regression model with training data
```

```python
model.fit(X_train, Y_train)
```

## Evaluation of the model
#### Based on the Accuracy Score

```python
##accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy on training data : ', training_data_accuracy)
```

```python
#accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy Score on test data : ', test_data_accuracy)
```
